# Project Performance Analysis

This Jupyter notebook demonstrates how to perform exploratory data analysis and build predictive models for a project management dataset. The dataset is synthetic and simulates project attributes such as planned and actual durations, budgets, team sizes, complexity and risk levels. We explore the relationships in the data, visualize distributions and correlations, and build both classification and regression models to predict project success and costs.


## Load Dataset

In this section we load the dataset and inspect its structure. The `start_date` column is parsed as a date and the first few rows are displayed.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error, r2_score
from sklearn.linear_model import LogisticRegression, LinearRegression

# Load dataset (ensure correct relative path)
df = pd.read_csv('data/synthetic_project_data.csv', parse_dates=['start_date'])

# Display first 5 rows
pd.set_option('display.max_columns', None)
df.head()
```

## Exploratory Data Analysis

We start by examining summary statistics and distribution plots for key variables. This helps identify central tendencies and potential outliers. We also calculate correlation coefficients between numeric features.

```
# Summary statistics for numeric columns
summary = df.describe()
summary
```

```
# Histogram of cost and schedule overruns
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df['cost_overrun_percent'], bins=30, kde=True, ax=axs[0], color='orange')
axs[0].set_title('Cost Overrun Percentage Distribution')
axs[0].set_xlabel('Cost Overrun %')
axs[0].set_ylabel('Frequency')

sns.histplot(df['schedule_delay_percent'], bins=30, kde=True, ax=axs[1], color='green')
axs[1].set_title('Schedule Delay Percentage Distribution')
axs[1].set_xlabel('Schedule Delay %')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```
# Correlation matrix heatmap
plt.figure(figsize=(10, 8))
correlation = df[['planned_duration_days','actual_duration_days','planned_budget','actual_cost','team_size','complexity','risk_level','change_requests','cost_overrun_percent','schedule_delay_percent','success']].corr()
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Classification Model: Predicting Project Success

We train a logistic regression model to predict whether a project is successful or not. Success is determined by both low cost overrun and low schedule delay. We evaluate the model on a hold‑out test set and print classification metrics.

```
# Features and target for classification
features = ['planned_duration_days','planned_budget','team_size','complexity','risk_level','change_requests']
X = df[features]
y = df['success']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train logistic regression model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

# Predictions and evaluation
y_pred = log_reg.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("
Classification Report:
", classification_report(y_test, y_pred))
```

## Regression Model: Predicting Project Costs

Next we build a linear regression model to predict the actual project cost based on the same feature set. The model performance is measured using Mean Squared Error (MSE) and R‑squared (\(R^2\)).

```
# Target for regression
y_cost = df['actual_cost']

# Train/test split (reuse X_train, X_test from classification for simplicity)

# Train linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_cost.loc[X_train.index])

# Predictions and evaluation
cost_pred = lin_reg.predict(X_test)
mse = mean_squared_error(y_cost.loc[X_test.index], cost_pred)
r2 = r2_score(y_cost.loc[X_test.index], cost_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

## Conclusion

This notebook illustrated how to perform a basic analysis on a project management dataset. Using simple models we predicted both project success (classification) and actual cost (regression). You can extend this analysis by engineering additional features (e.g. project type, geographic region) or by experimenting with more sophisticated algorithms such as decision trees, random forests or gradient boosting.
